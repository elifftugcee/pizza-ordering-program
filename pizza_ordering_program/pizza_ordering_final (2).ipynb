```python
#Create Pizza superclass and it's subclasses
class Pizza:
    def __init__(self,description,price):
        self._description= description
        self._price=price
    def get_description(self):
        return self._description
    def get_Cost(self):
        return self._price
```

```python
class Classic(Pizza):
    def __init__(self):
        description="Classic pizza includes tomato sauce, mozzarella, onion rings. "
        price=12.0
        super().__init__(description,price)
        
class Margarita(Pizza):
    def __init__(self):
        description="Margherita pizza includes tomatoes, mozzarella  cheese, fresh basil, salt, olive oil. "
        price=15.0
        super().__init__(description,price)

class Turkish_pizza(Pizza):
    def __init__(self):
        description="Turkish pizza thin base which is topped with a mixture of vegetables, meat and sauce."
        price=20.0
        super().__init__(description,price)
        
class Plain(Pizza):  
    def __init__(self):
        description="Plain pizza includes just cheese and sauce."
        price=10.0
        super().__init__(description,price)        
```

```python
#Here we capsullated price and description variables with get_description() and get_cost() methods.
#Create Decorator Class
#This decorator class is super class of sauce classes. Thanks to this class we will be able to add extra sauces.
class Decorator_pizza(Pizza):
    def __init__(self, pizza):
        self.component= pizza
    def get_description(self):
        return self.component.get_description() + Pizza.get_description(self)
    def get_Cost(self):
        return self.component.get_Cost()+ Pizza.get_Cost(self)           
```

```python
class Olive(Decorator_pizza):
    def __init__(self,pizza):
        self._description="Olive"
        self._price= 3.0
        super().__init__(pizza)

class Mushroom(Decorator_pizza):
    def __init__(self,pizza):
        self._description="Mushrooms"
        self._price= 4.0
        super().__init__(pizza)

class Goat_Cheese(Decorator_pizza):
    def __init__(self,pizza):
        self._description="Goat Cheese"
        self._price= 5.0
        super().__init__(pizza)

class Meat(Decorator_pizza):
    def __init__(self,pizza):
        self._description="Meat"
        self._price= 5.0
        super().__init__(pizza)        

class Onion(Decorator_pizza):
    def __init__(self,pizza):
        self._description="Onion"
        self._price= 1.0
        super().__init__(pizza) 

class Corn(Decorator_pizza):
    def __init__(self,pizza):
        self._description="Corn"
        self._price= 3.0
        super().__init__(pizza)
    
```

```python
#Get order from customer
def main():
    print("----------Welcome to Bny's Pizzas----------")
    with open("Menu.txt","r") as file:
        menu=file.read()
        print(menu)
    order =True
   
    while order:
        choice_pizza=input("Please choose pizza between 1-4 : \n")
        if choice_pizza=="1":
            choice_pizza=Classic()
            break
        elif choice_pizza=="2":
            choice_pizza= Margarita()
            break
        elif choice_pizza=="3":
            choice_pizza= Turkish_pizza() 
            break
        elif choice_pizza=="4":
            choice_pizza= Plain()
            break
        else:
            print("Wrong choice. Please try again...")
        #order=False  
    
    choice_sauces=[]
    while True:
        csauces= input("Please choose pizza between 5-10. When your selection process is complete click on 'finish'.: \n")
        if csauces=="5":
            choice_sauces.append(Olive(choice_pizza))
        elif csauces=="6":
            choice_sauces.append(Mushroom(choice_pizza))
        elif csauces=="7":
            choice_sauces.append(Goat_Cheese(choice_pizza))
        elif csauces=="8":
            choice_sauces.append(Meat(choice_pizza))
        elif csauces=="9":
            choice_sauces.append(Onion(choice_pizza))
        elif csauces=="10":
            choice_sauces.append(Corn(choice_pizza))  
        elif csauces=="finish":
            break
    choosed_Sauces=[]
    total_cost=choice_pizza.get_Cost()
    for sauce in choice_sauces:
        total_cost+=sauce.get_Cost()
        choosed_Sauces.append(sauce)
    print("Your pizza order includes:")
    for sauce in choosed_Sauces:
        print(sauce.get_description())
        
    print("Total cost:", total_cost)    
main()
```
