```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from pandas.api.types import CategoricalDtype
```

```python
sns.set(rc={'figure.figsize':(6,6)})
```

```python
df_airbnb = pd.read_csv('AB_NYC_2019.csv')
df_airbnb.head()
```

```python
df_airbnb["room_type"].value_counts().plot.barh().set_title("Class Frequencies of Room_type Variable")
```

```python
df_airbnb["room_type"].value_counts().plot.bar().set_title("Class Frequencies of Room_type Variable")
```

```python
df_airbnb["room_type"].index
```

```python
sns.barplot(x="room_type", y=df_airbnb["room_type"].index, data=df_airbnb, color="orange").set_title("Room Type")
```

```python
sns.barplot(x="room_type",y="price",hue="neighbourhood_group",data=df_airbnb)
plt.title("Room_Type - Price by Neighbourhood_Group")
```

```python
sns.barplot(x="room_type",
            y="price",
            hue="neighbourhood",
            data=df_airbnb)
plt.title("Room_Type - Price by Neighbourhood")
```

```python
df_airbnb.info()
```

```python
df_airbnb.groupby("room_type").sum().plot.pie(y="price", autopct="%.2f%%", ylabel="", legend=False, figsize=(6,4))
```

```python
cols = ["latitude", "longitude", "room_type"]
for col in cols:
  plt.figure(figsize=(6,6))
  sns.scatterplot(x=col, y="price", data=df_airbnb)
```

```python
sns.histplot(df_airbnb["price"], kde=True) # bins= auto
```

```python
sns.histplot(df_airbnb["price"], kde=True, bins=40)
```

```python
sns.kdeplot(df_airbnb["price"])
```

```python
sns.jointplot(x="price", y=df_airbnb["latitude"], data=df_airbnb)
```

```python
sns.catplot(x="room_type", y="price", data=df_airbnb)
plt.title("Room Type - Price")
plt.xlabel("Room Type")
plt.ylabel("Price")
plt.xticks(rotation=45)
```

```python
sns.catplot(x="room_type", y="price", hue="neighbourhood_group", kind="point", data=df_airbnb)
plt.title("Room Type - Price by Neighbourhood Group"),
plt.xticks(rotation=45)
```

```python
sns.kdeplot(df_airbnb["price"], shade=True)
```

```python
g = sns.FacetGrid(df_airbnb, hue="room_type", height=5, xlim=(0, 1000))
g.map(sns.kdeplot, "price", shade=True)
g.add_legend()
```

```python
plt.title("Correlation Matrix")
sns.heatmap(df_airbnb.corr(), annot=True);
```
